```python
import pandas as pd
import json
import numpy as np
```

```python
men_url = "http://115.78.93.252/TTU/pool/donghuynh0/data_visualization_2025/final_proejct/men_results.csv"
women_url = "http://115.78.93.252/TTU/pool/donghuynh0/data_visualization_2025/final_proejct/women_results.csv"
```

```python
men_df = pd.read_csv(men_url)
women_df = pd.read_csv(women_url)
```

```python
men_df = men_df.drop("Unnamed: 0", axis=1)
women_df = women_df.drop("Unnamed: 0", axis=1)
```

```python
with open('continent_mapping.json', 'r') as f:
    continent_mapping = json.load(f)
```

```python
continent_mapping
```

```python
men_df['home_continent'] = men_df['home_team'].map(continent_mapping)
men_df['away_continent'] = men_df['away_team'].map(continent_mapping)

women_df['home_continent'] = women_df['home_team'].map(continent_mapping)
women_df['away_continent'] = women_df['away_team'].map(continent_mapping)
```

```python
men_df['goal_difference'] = np.abs(men_df['home_score'] - men_df['away_score'])
men_df['winner'] = np.where(men_df['home_score'] > men_df['away_score'], men_df['home_team'],
                   np.where(men_df['away_score'] > men_df['home_score'], men_df['away_team'],
                   'Draw'))
```

```python
women_df['goal_difference'] = np.abs(women_df['home_score'] - women_df['away_score'])
women_df['winner'] = np.where(women_df['home_score'] > women_df['away_score'], women_df['home_team'],
                   np.where(women_df['away_score'] > women_df['home_score'], women_df['away_team'],
                   'Draw'))
```

```python
men_df['date'] = pd.to_datetime(men_df['date'])
men_df['year'] = men_df['date'].dt.year
```

```python
women_df['date'] = pd.to_datetime(women_df['date'])
women_df['year'] = women_df['date'].dt.year
```
